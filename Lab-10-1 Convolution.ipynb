```python
import torch
import torch.nn as nn
```

```python
# 연습1 - CNN 모델

conv = nn.Conv2d(1,1,11, stride=4, padding=0) # 입력채널, 출력채널, 커널 크기, padding, stride, bias=True 등
print(conv)

input = torch.Tensor(1,1,227,227) # (batch_size, channel, height, width)
print(input.shape)

output = conv(input)
print(output.shape)
```

```python
# 연습2 - pooling
# input이 홀수일때는?

input = torch.Tensor(1,1,11,11)
conv1 = nn.Conv2d(1,5,5)
pool = nn.MaxPool2d(2)

out = conv1(input)
out2 = pool(out)

print(out.size())
print(out2.size())
```
