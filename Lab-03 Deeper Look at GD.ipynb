```python
import torch
import numpy as np
```

```python
x_train = torch.FloatTensor([[1],[2],[3]])
y_train = torch.FloatTensor([[1],[2],[3]])
```

## Gradient Descent 를 직접 구현

```python
w = torch.zeros(1)
# b = torch.zeros(1, requires_grad=True)
```

```python
epochs = 10

for epoch in range(1, epochs+1):
    lr = 0.1
    hypothesis = x_train * w

    cost = torch.mean((hypothesis - y_train)**2)

    gradient = lr * torch.mean((hypothesis-y_train) * x_train)

    print("반복: {}, Cost: {}, W값: {}".format(epoch, cost, w))

    w -= gradient
```

## Gradient Descent with torch.optim

```python
w = torch.zeros(1, requires_grad = True)
```

```python
epochs = 10
optimizer = torch.optim.SGD([w], lr = 0.15)

for eapoch in range(1, epochs+1):
    hypothesis = x_train * w
    cost = torch.mean((hypothesis - y_train)**2)

    optimizer.zero_grad()
    cost.backward()
    optimizer.step()
```
